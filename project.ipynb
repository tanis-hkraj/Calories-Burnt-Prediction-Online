```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
df_exercise=pd.read_csv('exercise.csv')
df_calories=pd.read_csv('calories.csv')
```

```python
df_exercise.head(5)
```

```python
df_calories.head(5)
```

```python
df=pd.concat([df_exercise,df_calories['Calories']],axis=1)
df.head(5)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
```

```python
# statistical info of the Data
df.describe()
```

```python
# Finding Correlation
df.corr()
```

```python
df.head(5)
```

```python
X=df.drop(columns=['User_ID','Calories'])
y=df['Calories']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_test.info()
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred1=lr.predict(X_test)
print(metrics.r2_score(y_pred1,y_test))
```

```python
import pickle
with open("model.pkl","wb") as files:
    pickle.dump(lr,files)
```
